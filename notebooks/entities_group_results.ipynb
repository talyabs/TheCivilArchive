```python
entity_list = ['ORG', 'PERS_CLEAN', 'DATE_CLEAN', 'LOC']
# Function to count occurrences of each entity


def convert_to_list(column):
    return column.apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to count occurrences of each entity
def count_entities(column):
    # Convert the column elements to lists if they are strings
    column = convert_to_list(column)
    
    # Flatten the list of lists and count occurrences
    return Counter([item for sublist in column.dropna() if isinstance(sublist, list) for item in sublist])

# Assuming 'df' is your DataFrame and 'entity_list' contains the columns you want to process
entity_list = ['ORG', 'PERS_CLEAN', 'DATE_CLEAN', 'LOC']
entity_counts = {}

# Iterate over the specified columns and generate counts
for col in entity_list:
    counts = count_entities(df[col])
    entity_counts[col] = pd.DataFrame(counts.items(), columns=['name', 'count']).sort_values(by='count', ascending=False)

# To display the result
for key, count_df in entity_counts.items():
    print(f"Entity Counts for {key}:")
    print(count_df)
    print("\n")
```
